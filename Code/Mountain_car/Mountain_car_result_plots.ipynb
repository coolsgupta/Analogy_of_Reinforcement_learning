## Deep Q learning - Training data

```python
% matplotlib inline 

import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('results_DQN_model_2.csv')


#Storing the different columns in variables
episodes = df['episode']
epsilon = df['epsilon']
score = df['score']
average_score=df['average_score']
total_reward=df['total_reward']
average_reward=df['average_reward']


plt.title("DQN : Scores (Training)",fontsize=15)
plt.xlabel("Episode",fontsize=13)
plt.ylabel("Score",fontsize=13)
plt.plot(episodes,score,'c-')

#PLotting min and max values
ymax=max(score)
ymin=min(score)

#xmin
xmin = df.loc[df['score']==ymin]

if xmin.shape[0]>1:
    xmin=(xmin.iloc[[0]]) 
              

xmin = min(xmin['episode'])


#xmax
xmax = df.loc[df['score']==ymax]

if xmax.shape[0]>1:
    xmax=(xmax.iloc[[0]]) 
              
xmax=max(xmax['episode'])

#plotting min coordinates
plt.plot(xmin,ymin,'r.')
plt.text(xmin + 2,ymin,'min('+str(xmin)+','+str(ymin)+')',fontsize=11)

#plotting max coordinates
plt.plot(xmax,ymax,'r.')
plt.text(xmax + 2,ymax,'max('+str(xmax)+','+str(ymax)+')',fontsize=11)

plt.savefig('outputs/dql_score.png')
plt.show()

```

```python
#Average Scores
plt.title("DQN : Average scores (Training)",fontsize=15)
plt.xlabel("Episode",fontsize=13)
plt.ylabel("Average Score",fontsize=13)
plt.plot(episodes,average_score,'b-')

#PLotting min and max values
ymax=max(average_score)
ymin=min(average_score)

#xmin
xmin = df.loc[df['average_score']==ymin]

if xmin.shape[0]>1:
    xmin=(xmin.iloc[[0]]) 
              

xmin = min(xmin['episode'])


#xmax
xmax = df.loc[df['average_score']==ymax]

if xmax.shape[0]>1:
    xmax=(xmax.iloc[[0]]) 
              
xmax=max(xmax['episode'])

#plotting min coordinates
plt.plot(xmin,ymin,'r.')
plt.text(xmin + 2,ymin,'min('+str(xmin)+','+str(round(ymin,2))+')',fontsize=11)

#plotting max coordinates
plt.plot(xmax,ymax,'r.')
plt.text(xmax + 2,ymax,'max('+str(xmax)+','+str(round(ymax,2))+')',fontsize=11)


plt.savefig('outputs/dql_avg_score.png')
plt.show()
```
